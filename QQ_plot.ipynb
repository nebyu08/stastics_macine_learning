```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats
```

```python
n=10001
#data=np.random.randn(n)
data=np.exp(np.random.randn(n)*.8)

x=np.linspace(-4,4,1000)

prob_density=stats.norm.pdf(x)
norm_prob=prob_density/max(prob_density) #let normalize

y1,x1=np.histogram(data,bins=55)
x1=(x1[1:]+x1[:-1])/2
y1=y1/max(y1) #lets normalize

plt.plot(x1,y1,label="normal distribution")
plt.plot(x,norm_prob,label="probability distribution/theoretical")
plt.legend()
plt.show()
```

# lets creaete a QQ PLOT 

```python
# my version of QQ
z_sorted=np.sort(stats.zscore(data))
norm_sorted=stats.norm.ppf(np.linspace(0,1,n))
plt.plot(z_sorted,norm_sorted);
```

```python
# lets see them individually 
#fig,ax=plt.subplots(2,1,figsize=(8,9))
#ax[0].plot(z_sorted,'o')
#ax[1].plot(norm_sorted,'o')
plt.plot(norm_sorted,z_sorted,'o')
#lims
xl,xr=plt.xlim()
yl,yr=plt.ylim()
lim=[np.min([xl,xr,yl,yr]),np.max([xl,xr,yl,yr])]

plt.xlim(lim)
plt.ylim(lim)

plt.plot(lim,lim)
plt.xlabel("theoretical data")
plt.ylabel("observed data")
plt.title("QQ-plot")
plt.show()
```
