```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
```

```python
np.random.seed(42)
```

```python
# lets make some distributions
n=1000
nbins=20

d1=np.random.randn(n)-3
d2=2*np.random.randn(n)
d3=np.random.randn(n)+2

#first histogram
y1,x1=np.histogram(d1,nbins)
x1=(x1[1:]+x1[:-1])/2

#second histogram
y2,x2=np.histogram(d2,nbins)
x2=(x2[1:]+x2[:-1])/2

#third histogram
y3,x3=np.histogram(d3,nbins)
x3=(x3[1:]+x3[:-1])/2

#lets draw the histogram
plt.plot(x1,y1)
plt.plot(x2,y2)
plt.plot(x3,y3)
plt.title("the normal distribution")

plt.xlabel("data value")
plt.ylabel("data counts")

plt.show()
```

```python
# overlay the mean

#computing the mean of the distributions
mean_d1=np.mean(d1)
mean_d2=np.mean(d2)
mean_d3=np.mean(d3)

#lets plot the mean on the plot
plt.plot(x1,y1,'b',x2,y2,'r',x3,y3,'k')
plt.plot([mean_d1,mean_d1],[0,max(y1)],'b--')
plt.plot([mean_d2,mean_d2],[0,max(y2)],'r--')
plt.plot([mean_d3,mean_d3],[0,max(y3)],'k--')
plt.title("over lay mean ditribution")
plt.xlabel("data values")
plt.ylabel("data counts")
plt.show()
```

```python
# faiure of teh mean
nbins=24
n=2000
da=np.hstack((np.random.randn(n)+2,np.random.randn(n)-2))
yy,xx=np.histogram(da,nbins)
xx=(xx[1:]+xx[:-1])/2
plt.plot(xx,yy,'b')
plt.plot([np.mean(xx),np.mean(xx)],[0,max(yy)],'--b')
plt.title("failure of 'mean'")
plt.show()
```

```python
#my median...of the log
shift=0
streach=.7
n=2000
nbins=25

data=streach*np.random.randn(n)+shift
data=np.exp(data)


yy1,xx1=np.histogram(data,nbins)
xx1=(xx1[1:]+xx1[:-1])/2
log_mean=np.mean(data)
log_median=np.median(data)

#plot the median and the mean on the log 
fig,ax=plt.subplots(2,1,figsize=(5,8))
ax[0].plot(data,".",label="data")
ax[0].plot([1,n],[log_mean,log_mean],'r--',label="mean")
ax[0].plot([1,n],[log_median,log_median],'b--',label="median")
ax[0].legend()

ax[1].plot(xx1,yy1)
ax[1].plot([log_mean,log_mean],[0,max(yy1)],'r--',label="mean")
ax[1].plot([log_median,log_median],[0,max(yy1)],'b--',label="median")
plt.show()
```
