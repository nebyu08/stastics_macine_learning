```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats
```

```python
n=1000
data=np.random.randn(n)**2
plt.plot(data,'p')
```

```python
# lets calculate the first and third quartile...

ranked_data=stats.rankdata(data)/n #lets rank and normalize the data

q1=np.argmin((ranked_data-.25)**2)
q3=np.argmin((ranked_data-.75)**2)

#lets get the values
quar_val=data[[q1,q3]]

#the interquartile is the vaues between them
iq_range=quar_val[1]-quar_val[0]

#or use the inter-quartile value
iq_range_2=stats.iqr(data)
```

```python
print(iq_range,iq_range_2)
```
