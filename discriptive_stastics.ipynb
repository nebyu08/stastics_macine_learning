```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats
```

```python
n=1001 # this is the number of descitization

x=np.linspace(-4,4,n)
gaudist=stats.norm.pdf(x)

plt.plot(x,gaudist)
plt.title("analytics gausian data distribution")
plt.show()
```

```python
np.sum(gaudist)
```

```python
# lets make a some what of probability distribution
plt.plot(x,gaudist/sum(gaudist))
plt.title("the ANALYTIC probabiity gausian distribution")
print(sum(gaudist)/sum(gaudist))
#the aread under a curve is
print(sum(gaudist)*np.diff(x[:2]))
```

```python
# lets make a Empirical normal distribution
shift=1  #shift
streach=5 #standard deviation
n=1000

data=streach*np.random.randn(n)+shift
plt.hist(data,25)
plt.title("Emperical normal distribution")
plt.show()
print(np.mean(data))
```

```python
# uniformaly distributed 
fig,ax=plt.subplots(2,1,figsize=(5,8))

stretch=2
shift=0.5
n=1000
data=stretch*np.random.rand(n)+shift-stretch/2
#histogram
ax[0].hist(data,25)
ax[0].set_title("uniformay distributed histogram")

#noraml graph plot
ax[1].plot(data,'.',markersize=5)
ax[1].set_title("plot of uniformally ditributed")
plt.show()
print("standard deviation is:",np.std(data))
```

```python
# log normal distribution
n=1000
x=np.linspace(0,10,n)
data=stats.lognorm.pdf(x,1)
plt.plot(x,data)
plt.title("analytical log normal distribution")
plt.plot()
```

```python
# emperical normal distribution
n=2000
shift=5
streach=0.5
data=streach*np.random.randn(n)+shift
data=np.exp(data)

fig,ax=plt.subplots(2,1,figsize=(5,8))
ax[0].plot(data,".")
ax[0].set_title("emperical log normal distribution")

ax[1].hist(data,25)
ax[1].set_title("emperical log noraml distribution")

plt.show()
print("standard deviation is:",1/np.exp(np.std(data)))
```

```python
#binomial probabiity distribution
p=0.5 # the probability of that happening
t=10 # the total number of distribution
x=range(t+2)

bindist=stats.binom.pmf(x,t,p)

plt.bar(x,bindist)
plt.title("the binomial probability distribution (%d toss and probabiity %f)"%(t,p))
plt.show()
```

```python
# T-DISTRIBUTION
x=np.linspace(-4,4,1000)
df=45
data=stats.t.pdf(x,df)

plt.plot(data)
plt.title("T-distribution")
plt.show()
```

```python
# THE F-DISTRIBUTION
num_df=12 # numinator degree of freedom
den_df=100 #denominator degree of freedom

x=np.linspace(-4,4,1000)
fdist=stats.f.pdf(x,num_df,den_df)

plt.plot(x,fdist)
plt.title("F-DISTRIBUTION")
plt.plot()
```
